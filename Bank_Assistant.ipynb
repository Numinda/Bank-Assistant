```python
import random

# Define a list of possible user greetings and bot responses.
greetings = ['hello', 'hi', 'hey', 'hola', 'greetings']
responses = ['Hello! How can I assist you today?', 'Hi there! How may I help you?', 'Hey! What can I do for you?']

# Define a list of common user questions and corresponding answers.
questions = ['what is my account balance', 'how to transfer money', 'help', 'bye']
answers = [
    'Your account balance is $XXXX.',
    'To transfer money, you can use our mobile app or visit a branch.',
    'Sure, I can help you with that. What do you need assistance with?',
    'Goodbye! If you have more questions, feel free to ask.'
]

# Function to generate responses based on user input.
def chatbot_response(user_input):
    user_input = user_input.lower()
    if user_input in greetings:
        return random.choice(responses)
    elif user_input in questions:
        return answers[questions.index(user_input)]
    else:
        return "I'm sorry, I couldn't understand your request. Please ask another question or type 'help'."

# Main chat loop.
print("AI Bank Chatbot: Hi there! How may I assist you? (Type 'bye' to exit)")
while True:
    user_input = input("You: ")
    if user_input.lower() == 'bye':
        print("AI Bank Chatbot: Goodbye! Have a great day.")
        break
    response = chatbot_response(user_input)
    print("AI Bank Chatbot:",response)
```
